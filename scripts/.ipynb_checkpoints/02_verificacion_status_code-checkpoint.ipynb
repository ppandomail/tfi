# Script de verificación de éxito de la solicitud (request)
---

```python
import requests
import pandas as pd
```

```python
df = pd.read_csv('universidades.csv', sep=';', header=None)
```

```python
colname = df.columns[1]

for valor in df[colname]:
    try:
        print(valor, '->', requests.get(valor).status_code)
    except Exception as e:
        print(valor, '->', e)
```
