# Script de verificación de éxito de la solicitud (request)
---

```python
import requests
import pandas as pd
```

```python
df = pd.read_csv('universidades-nombre-url.csv', sep=';')
```

```python
df.head()
```

```python
# Función para obtener el status_code de cada URL
def get_status(url):
    try:
        response = requests.get(url, verify=False, timeout=5)
        return response.status_code
    except requests.exceptions.RequestException:
        return None  # Si hay error, devuelve None
```

```python
df['status_code'] = df['URL'].apply(lambda url: get_status(url))
```

```python
df.to_csv('universidades-nombre-url-code.csv', sep=';', index=False)
```
