# Script de extracción de urls de universidades argentinas mediante web scraping
---

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO
```

```python
url = 'https://guiadecarreras.siu.edu.ar/ciie_ofertas/2.0/listado_instituciones.php'
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
tables = soup.find_all('table')
```

```python
df = pd.read_html(StringIO(str(tables)))[0]
```

```python
df[1] = df[1].apply(lambda url: f'https://{url}')
```

```python
df.head()
```

```python
df.to_csv('universidades-nombre-url.csv', sep=';', index=False, header=False)
```
